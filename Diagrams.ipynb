```python
import json
```

```python
file = open('BERT-NER-TF2/log/results-tiny-nontrained-conll.json')
data = json.load(file)

tiny_nontrained_conll=[]
for i in data['results']:
    # print(i['macro avg'])
    # print(i['macro avg']['f1-score'])
    tiny_nontrained_conll.append(i['macro avg']['f1-score'])

file.close()

# print(len(tiny_nontrained_conll))
print(tiny_nontrained_conll)
```

```python
file = open('BERT-NER-TF2/log/results-tiny-pretrained-conll.json')
data = json.load(file)

tiny_pretrained_conll=[]
for i in data['results']:
    # print(i['macro avg'])
    # print(i['macro avg']['f1-score'])
    tiny_pretrained_conll.append(i['macro avg']['f1-score'])

file.close()

# print(len(tiny_pretrained_conll))
print(tiny_pretrained_conll)
```

```python
file = open('BERT-NER-TF2/log/results-tiny-pretrained-few_nerd.json')
data = json.load(file)

tiny_pretrained_fewnerd=[]
for i in data['results']:
    # print(i['macro avg'])
    # print(i['macro avg']['f1-score'])
    tiny_pretrained_fewnerd.append(i['macro avg']['f1-score'])

file.close()

# print(len(tiny_pretrained_fewnerd))
print(tiny_pretrained_fewnerd)
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

matplotlib.rcParams['font.sans-serif'] = "Times New Roman"
matplotlib.rcParams['font.family'] = "serif"
```

```python
rounds10 = range(1,51)
plt.plot(rounds10, tiny_nontrained_conll, 'g', label='Nontrained')
plt.plot(rounds10, tiny_pretrained_conll, 'b', label='Pretrained')
# plt.title('Training TINY model with CoNLL-2003 average macro f1-score')
plt.xlabel('Rounds 10x')
plt.ylabel('Average macro f1-score')
plt.legend()
fig1 = plt.gcf()
plt.show()
plt.draw()
```

```python
fig1.savefig('fig-tiny-conll.png', dpi=180, transparent=True)
```
